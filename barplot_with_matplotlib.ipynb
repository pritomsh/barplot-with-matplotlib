# Barplot
- A barplot shows the relationship between a numeric and a categoric variable. Each entity of the categoric variable is represented as a bar. The size of the bar represents its numeric value. 

```python
# Libraries
import matplotlib.pyplot as plt
import pandas as pd
data = pd.read_csv("data/supershops.csv")
data.head()
```

```python
data.Area.unique()
```

```python
data.Area.value_counts()
```

```python
# Create bars
plt.figure(figsize=(5,4))

plt.bar(data.Area.unique(), data.Area.value_counts(), width=.3,
        color=['black', 'red', 'green', 'blue', 'cyan']) #change each bar colors  

# Add title
plt.title('A simple barplot')

# Show graphic
plt.show()
```

Plotting a Horizontal Barplot using Matplotlib

```python
# Create horizontal bars
plt.barh(data.Area.unique(), data.Area.value_counts(), color=(0.2, 0.6, 0.6, 0.6))
# Add title
plt.title('A horizontal barplot');
# Show graphic
plt.show()
```

# Add Text watermark

```python
# Create bars
fig = plt.figure(figsize=(5,4))

plt.bar(data.Area.unique(), data.Area.value_counts(), width=.3,
        color=['black', 'red', 'green', 'blue', 'cyan']) #change each bar colors  
plt.xticks(rotation=45)

# Add title
plt.title('A Text watermark in barplot')

# Add Text watermark
fig.text(.6, -0.15, ' Pritom Saha', fontsize = 12,
         color ='red', ha ='right', va ='bottom',
         alpha = 0.7)
# Show graphic
plt.show()
```

```python
#Control color of border
#The edgecolor argument allows you to color the borders of barplots.



# Create bars with blue edge color
plt.bar(data.Area.unique(), data.Area.value_counts(), color=['green'],  edgecolor='red')
# Add title
plt.title('A color of border and edgecolor')
# Show graph
plt.show()
```

```python
'''In the following example, title, x label and y label are added to the barplot using the title(), xlabel(), 
and ylabel() functions of the matplotlib library.

Those functions are applied to a barplot in the example, but the same method would work for other chart types.'''


 
# Create bars and choose color
plt.bar(data.Area.unique(), data.Area.value_counts(), color = (0.5,0.1,0.5,0.6))
 
# Add title and axis names
plt.title('Area vs Values')
plt.xlabel('Area')
plt.ylabel('Values')
 

 
# Show graph
plt.show()
```

```python
#In the following example, title, x label and y label are added to the barplot using the title(), xlabel(), and ylabel() functions of the matplotlib library.
#Those functions are applied to a barplot in the example, but the same method would work for other chart types.
plt.style.use('seaborn-darkgrid')

# Create bars and choose color
plt.bar(data.Area.unique(), data.Area.value_counts(), color = (0.5,0.1,0.5,0.6))
 
# Add title and axis names
plt.title('Category vs Values')
# plt.xlabel('categories')
plt.xlabel('category', fontweight='bold', color = 'green', fontsize='18')

plt.ylabel('values', color = 'red', fontsize='18')

# Create names on the X-axis
plt.xticks(color='orange')
# Create names on the Y-axis
plt.yticks(color='orange')

 
# Show graph
plt.show()
```
