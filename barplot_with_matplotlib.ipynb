```python
# Libraries
import matplotlib.pyplot as plt
import pandas as pd
data = pd.read_csv("data/supershops.csv")
data.head()
```

# Data Preprocessing

```python
#check null values 
print("Check is there any null values there")
print(data.isnull().sum())
```

```python
data = data.dropna()
print("After Drop Null values")
print(data.isnull().sum())
```

```python
# check all columns name
print("Feature Names")
print(data.columns)
```

```python
# print all each columns unique category
print(f"Unique Values for categorical columns:")
for col in data.select_dtypes(include='object'):
  print(f"  - {col}: {data[col].unique()}\n")
```

```python
print("Values of Each Labels in Area \n",)
print(data.Area.value_counts())
```

# make new dataset for show in chart

```python
newdata = data.groupby('Area').size().rename_axis('Area').reset_index(name='Count')
newdata.head()
labels = newdata.Area
values = newdata.Count
```

# Visualization

```python
# Create bars
plt.figure(figsize=(5,4))
plt.bar(labels, values)  
# Show graphic
plt.show()
```

# Add title

```python
# Create bars
plt.figure(figsize=(5,4))
plt.bar(labels, values) 
plt.title('This is barplot');
# Show graphic
plt.show()
```

#  Horizontal Barplot 

```python
#  For Create horizontal bars just use plt.barh()
plt.figure(figsize=(5,4))
plt.barh(data.Area.unique(), data.Area.value_counts(), color=(0.2, 0.6, 0.6, 0.6))
# Add title
plt.title('A horizontal barplot');
# Show graphic
plt.show()
```

# Custom Color

```python
#create color set
color = ['lightcoral', 'rosybrown', 'sandybrown', 'navajowhite', 'gold',
        'khaki', 'lightskyblue', 'turquoise', 'lightslategrey', 'thistle', 'pink'] 

# Create bars
plt.figure(figsize=(5,4))
plt.bar(labels, values, color = color) 
plt.title('This is barplot');
# Show graphic
plt.show()
```

#  Control width and space

```python
#create color set
color = ['lightcoral', 'rosybrown', 'sandybrown', 'navajowhite', 'gold',
        'khaki', 'lightskyblue', 'turquoise', 'lightslategrey', 'thistle', 'pink'] 

width = [0.8,0.5,0.5]
x_pos =[0,1,2]

# Create bars
plt.figure(figsize=(5,4))
plt.bar(labels, values, color = color, width= width) 

# Create names on the x-axis
plt.xticks(x_pos)
plt.title('This is barplot');
# Show graphic
plt.show()
```

# Add Text watermark

```python
# Create bars
fig = plt.figure(figsize=(5,4))

plt.bar(data.Area.unique(), data.Area.value_counts(), width=.3,
        color=['black', 'red', 'green', 'blue', 'cyan']) #change each bar colors  
plt.xticks(rotation=45)

# Add title
plt.title('A Text watermark in barplot')

# Add Text watermark
fig.text(.6, -0.15, '© Pritom Saha', fontsize = 12,
         color ='red', ha ='right', va ='bottom',
         alpha = 0.7)
# Show graphic
plt.show()
```

# Control color of border

- The edgecolor argument allows you to color the borders of barplots.

```python

plt.figure(figsize=(5,4))

# Create bars with blue edge color
plt.bar(labels, values, color=['navajowhite'],  edgecolor='red')

# Add title
plt.title('A color of border and edgecolor')
# Show graph
plt.show()
```

# x label and y label

```python
'''In the following example x label and y label are added to the barplot using the  xlabel(), 
and ylabel() functions of the matplotlib library.
'''

plt.figure(figsize=(5,4))
 
# Create bars and choose color
plt.bar(labels, values, color = (0.5,0.1,0.5,0.6))
 
# Add title and axis names
plt.title('Area vs Values')
plt.xlabel('Area')
plt.ylabel('Values')
 
# Show graph
plt.show()
```

# change color, fontweight, fontsize of  x label and y label

```python
plt.figure(figsize=(5,4))
 
# Create bars and choose color
plt.bar(labels, values, color = (0.5,0.1,0.5,0.6))
 
# Add title 
plt.title('Area vs Values', fontweight='bold')


# plt.xlabel('categories')
plt.xlabel('Area', fontweight='bold', color = 'green', fontsize='12')
plt.ylabel('values', color = 'red', fontsize='10')

# Create names on the X-axis
plt.xticks(color='orange')
# Create names on the Y-axis
plt.yticks(color='orange')

 
# Show graph
plt.show()
```

#  add value labels on top a bar chart

```python
plt.figure(figsize=(5,4))
# create a barplot veriable 
barplot1 = plt.bar(labels,values, color = (0.5,0.1,0.5,0.6))
# add labels = values and for top label_type = "edge"
plt.bar_label(barplot1, labels = values, label_type="edge")
plt.show()
```

#  add value labels on center a bar chart

For this just change label_type = "center"

```python
plt.figure(figsize=(5,4))
# create a barplot veriable 
barplot1 = plt.bar(labels,values, color = (0.5,0.1,0.5,0.6))
# add labels = values and for top label_type = "edge"
plt.bar_label(barplot1, labels = values, label_type="center")
plt.show()
```

# Change Style of plot

```python
plt.style.use('seaborn-notebook')
plt.figure(figsize=(5,4))
# Create bars and choose color
plt.bar(labels, values, color = (0.5,0.1,0.5,0.6))
 
# Add title and axis names
plt.title('Category vs Values')
# plt.xlabel('categories')
plt.xlabel('category', fontweight='bold', color = 'green', fontsize='18')

plt.ylabel('values', color = 'red', fontsize='18')

# Create names on the X-axis
plt.xticks(color='orange')
# Create names on the Y-axis
plt.yticks(color='orange')

 
# Show graph
plt.show()
```
